```python
import numpy as np
import pandas as pd
import sqlite3

from IPython.core.display import SVG

database = "database.sqlite"
```

```python
connection = sqlite3.connect(database)
```

```python
sql = """
        SELECT * 
        FROM sqlite_master
        WHERE type='table';
        """

table = pd.read_sql(sql, connection)

table
```

```python
sql = """
        SELECT
            C.name AS country_name,
            L.name AS league_name
        FROM Country AS C
        INNER JOIN League AS L
        ON C.id == L.id
        """
        
table = pd.read_sql(sql, connection)

table
```
