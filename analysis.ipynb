```python
sql = """
        SELECT 
            c.name as country_name,
            l.name as league_name,
            m.season,
            CASE WHEN m.home_team_goal > m.away_team_goal THEN ht.team_long_name WHEN m.home_team_goal < m.away_team_goal THEN at.team_long_name ELSE 'Draw' END as team_name,
            COUNT(CASE WHEN m.home_team_goal > m.away_team_goal THEN ht.team_long_name WHEN m.home_team_goal < m.away_team_goal THEN at.team_long_name ELSE 'Draw' END) as wins
        FROM match as m
        INNER JOIN country as c ON m.country_id = c.id
        INNER JOIN league as l ON m.league_id = l.id
        INNER JOIN team as ht ON m.home_team_api_id = ht.team_api_id
        INNER JOIN team as at ON m.away_team_api_id = at.team_api_id
        GROUP BY c.name, l.name, m.season, team_name
        HAVING team_name != 'Draw'
        ORDER BY wins DESC;
        """

table = pd.read_sql(sql, connection)

table
```
